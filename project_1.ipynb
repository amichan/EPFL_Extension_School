## B.1

```python
#1. Compute the mean happiness score for each region and rank the regions from most happy to least happy.
df_region=df.groupby('Region').mean()
```

```python
df_region.sort_values(['Happiness Score'],ascending=False, inplace=True)
```

```python
df_region[['Happiness Score']]
```

## B.2

```python
#2. Compute the number of countries in each region that have a happiness score above 6.0.
df_region['Countries_Happy']=np.arange(len(df_region))
for region in regions:
    value=(df[df['Region']== region ]['Happiness Score']>6.0).sum()
    df_region.loc[region, 'Countries_Happy'] = value
df_region[['Happiness Score','Countries_Happy']]
```

## B.3

```python
#3. Compute the difference between the maximum and minimum happiness 
#score for each region. Which region has the largest range of happiness scores?
df_region['Countries_Happy_Range']=np.arange(len(df_region))
for region in regions:
    max_val=(df[df['Region']== region ]['Happiness Score']).max()
    min_val=(df[df['Region']== region ]['Happiness Score']).min()
    df_region.loc[region, 'Countries_Happy_Range'] = max_val-min_val
df_region[['Happiness Score','Countries_Happy','Countries_Happy_Range']]
```

### The region with the highest range in happiness scores is Asia-Pacific.

## G. Probabilities

```python
# Compute the probability that randomly selected country with Happiness score over 6.0 is from Western Europe. 
# You will have to use pandas to count the appropriate quantities.
import random
experiments = 500000
counter = 0
df2=df[df['Happiness Score'] > 6.0]
for i in range(experiments):
    random_country = random.randrange(len(df2))
    if df2.iloc[random_country]['Region']=='Western Europe':
        counter += 1
counter/experiments
    
```

#### The probability is ~0.386

## H. Matrices

```python
# Define a matrix whose rows correspond to countries and the columns to the regions. Fill in the matrix with 
# 0/1 values where entry (i,j) is a 1 if the country in row i is in the region in column j and a 0 otherwise.

regions=columns=df['Region'].unique() #redundant from above, included for completeness/clarity
countries=df.index
df_bool = pd.DataFrame(index=countries, columns=regions)
```

```python
for i in countries:
    for j in regions:
        if df.loc[i]['Region']==j:
            df_bool.loc[i][j]=1
        else: 
            df_bool.loc[i][j]=0
            
```
